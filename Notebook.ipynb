```python
from IPython.display import Audio, Image, display
import time

def Open_AI(message, mode='text'):
    if mode == 'text':
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {
                    "role": "user",
                    "content": "{}".format(message)
                },
            ],
            temperature=1,
            max_tokens=256,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0
        )
        return response.choices[0].message.content
    elif mode == 'image':
        response = client.images.generate(
            model="dall-e-3",
            prompt="{}".format(message),
            size="1024x1024",
            quality="standard",
            n=1,
        )
        return response.data[0].url
    else:
        return "Invalid mode. Please use 'text' or 'image'."

def Open_AI2(message):
    response = client.audio.speech.create(
        model="tts-1",
        voice="shimmer",
        input=message
    )
    return response

while True:
    user_input = input("Enter a prompt (type 'quit' to exit): ")

    if user_input.lower() == 'quit':
        print("Exiting...")
        break

    mode_input = input("Enter mode ('text' or 'image'): ")

    if mode_input.lower() == 'text' or mode_input.lower() == 'image':
        
        if mode_input.lower() == 'text':
            print("Peppa Pig (text): ", Open_AI(user_input, mode='text'))
            chatbot_response = Open_AI(user_input)
    
            timestamp = str(int(time.time()))
            audio_file_path = f"C:/Users/Semal/Desktop/OpenAI Project/speech_output_{timestamp}.mp3"
    
            speech_response = Open_AI2(chatbot_response)
            speech_response.stream_to_file(audio_file_path)
    
            print("Playing audio...")
            audio = Audio(filename=audio_file_path, autoplay=True)
            display(audio)
            
        else:
            img_url = Open_AI(user_input, mode='image')
            print("Generated image URL:", img_url)
            display(Image(url=img_url))
    else:
        print("Invalid mode input. Please enter 'text' or 'image'.")

    print() 
```
